```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# the code to show the difference between image

```python
def calculate_difference(a,b):
    img = cv2.imread(a)
    img2 = cv2.imread(b)
    img2 = cv2.resize(img2,(img.shape[1],img.shape[0]))
    difference = cv2.absdiff(img, img2)
    total_diff = np.sum(difference)
    avg_diff = total_diff/(img.shape[1]*img.shape[0])
    print(total_diff, avg_diff)
    plt.imshow(img2-img)
    plt.colorbar()
    plt.show()
calculate_difference("ourcorner.png","ourcorner2.png")
```
