```python
import matplotlib.pyplot as plt
import cv2
```

# the code to show the difference between image

```python
img = cv2.imread("resources/data/normal.png")
img2 = cv2.imread("resources/data/test_data_1(6_points)-normal.png")
img2 = cv2.resize(img2,(img.shape[1],img.shape[0]))
plt.imshow(img2-img)

plt.colorbar()
plt.show()
```
